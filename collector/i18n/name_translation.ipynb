```python
# fill talent names
import json
data = json.load(open('eng_names.json'))
chn_data = json.load(open('chn_card.json'))
for key in data.keys():
    if 'TALENT_' in key:
        char_name = key.split('/')[0].split('_')[-1]
        real_name = key.split('/')[-1]
        char_chn_name = data[f'CHARACTOR/{char_name}']
        chn_name = ''
        for chn_key in chn_data.keys():
            if char_chn_name in chn_data[chn_key]:
                assert chn_name == ''
                chn_name = chn_key
        assert chn_name != ''
        # print(f'{key} {char_name} {chn_name}')
        print(f'"{key}": "{chn_name}",')
```

```python
# fill skills
type_dict = {
    'ELEMENTAL_SKILL': '元素战技',
    'ELEMENTAL_BURST': '元素爆发',
    'NORMAL_ATTACK': '普通攻击',
    'PASSIVE': '被动技能'
}
import json
data = json.load(open('eng_names.json'))
chn_data = json.load(open('chn_char.json'))
for key in data.keys():
    if 'SKILL_' in key:
        if 'Mob' in key:
            continue
        char_name = key.split('/')[0].split('_')[1]
        skill_type = '_'.join(key.split('/')[0].split('_')[2:])
        real_name = key.split('/')[-1]
        char_chn_name = data[f'CHARACTOR/{char_name}']
        skill_lists = chn_data[char_chn_name]
        found = []
        for skill in skill_lists:
            if skill[1] == type_dict[skill_type]:
                # assert len(found) == 0, (found, skill)
                found.append(skill)
        if len(found) != 1:
            print(f'"{key}": ""')
            continue
        else:
            found = found[0]

        # print(f'{key} {char_name} {chn_name}')
        print(f'"{key}": "{found[0]}",')
```

```python
# names to descs
import json
name_data = json.load(open('eng_names.json'))
desc_data = json.load(open('eng_descs.json'))
desc_keys = list(desc_data.keys())
chn_card = json.load(open('chn_card.json'))
chn_char = json.load(open('chn_char.json'))
chn_extra = json.load(open('chn_extra.json'))
items = list(name_data.items())
items.sort()
result = {}
for key in desc_data.keys():
    result[key] = ''
# print_sys = print
# def print(str):
#     key, value = str.split('": "')
#     key = key[1:].replace('"', r'\"').replace('\n', r'\n')
#     value = value[:-2].replace('"', r'\"').replace('\n', r'\n')
#     print_sys(f'"{key}": "{value}",')

def set_value(d, ks, v):
    if not isinstance(ks, list | tuple):
        ks = [ks]
    for k in ks:
        d[k] = v

for name, chn_name in items:
    real_name = name.split('/')[-1]
    main_type = name.split('/')[0].split('_')[0]
    if '_STATUS' in name:
        main_type = '_'.join(name.split('/')[0].split('_')[:2])
    desc_names = [x for x in desc_keys if f'{name}/' in x]
    if len(desc_names) == 0:
        # not found, maybe can find same name card or weapon or artifact
        print('no name found', chn_name, name)
        continue
    elif len(desc_names) != 1:
        descs = [desc_data[x] for x in desc_names]
        if len(set(descs)) != 1:
            # desc not all same, keep latest version
            latest = max(desc_names)  # keep latest version
            version = latest.split('/')[-1]
            print(f'multiple name found and desc different, keep version '
                  f'{version}', chn_name, name, desc_names)
            desc_names = [latest]
    else:
        pass
    card_types = ['ARCANE', 'CARD', 'ARTIFACT', 'WEAPON', 'SUPPORT', 'TALENT']
    if main_type in card_types:
        # is a card
        if chn_name not in chn_card:
            # print(f'"{desc_name}": "",')
            print('card name not found', name, chn_name)
            continue
        # print(f'"{desc_name}": "{chn_card[chn_name]}",')
        set_value(result, desc_names, chn_card[chn_name])
    elif main_type in ['CHARACTOR_STATUS', 'TEAM_STATUS', 'SUMMON']:
        # is extra
        chn_type = {
            'CHARACTOR_STATUS': '状态',
            'TEAM_STATUS': '出战状态',
            'SUMMON': '召唤物'
        }[main_type]
        target_dict = chn_extra[chn_type]
        if chn_name not in target_dict:
            # check if has same name card, weapon and artifact
            for tg in card_types:
                tg_name = f'{tg}/{real_name}'
                if tg_name in name_data:
                    assert name_data[tg_name] == chn_name
                    print(' extra name not found, but found same name in '
                          f'{tg}', name, chn_name)
                    set_value(result, desc_names, chn_card[chn_name])
                    break
            else:
                print('extra name not found', name, chn_name)
            continue
        # print(f'"{desc_name}": "{target_dict[chn_name]}",')
        set_value(result, desc_names, target_dict[chn_name])
    elif main_type == 'CHARACTOR':
        # is charactor, skip desc
        # print(f'"{desc_name}": "",')
        ...
    elif main_type == 'SKILL':
        # is skill
        charactor_name = name.split('/')[0].split('_')[1]
        chn_charactor_name = name_data[f'CHARACTOR/{charactor_name}']
        if '元素法师' in chn_charactor_name or '元素路人' in chn_charactor_name:
            chn_charactor_name = chn_charactor_name[-4:]
            skills = chn_char[chn_charactor_name]
            for sname, stype, sdesc in skills:
                if sname[-4:] == chn_name[-4:]:
                    set_value(result, desc_names, sdesc.replace('X', chn_name[0]))
                    # print(desc_data[desc_name], chn_name, sname, stype, sdesc)
                    break
            continue
        if chn_charactor_name not in chn_char:
            # print(f'"{desc_name}": "",')
            print('charactor name not found', name, chn_charactor_name)
            continue
        skills = chn_char[chn_charactor_name]
        for sname, stype, sdesc in skills:
            if sname == chn_name:
                # print(f'"{desc_name}": "{sdesc}",')
                set_value(result, desc_names, sdesc)
                break
        else:
            # print(f'"{desc_name}": "",')
            print('skill name not found', name, chn_name)
    elif main_type == 'EMPTY':
        # system handlers, skip desc
        # print(f'"{desc_name}": "",')
        ...
    else:
        raise NotImplementedError(main_type)
json.dump(result, open('chn_descs.json', 'w', encoding='utf-8'), 
          indent = 2, ensure_ascii = False)
```
