```python
import numpy as np
from PIL import Image
from IPython.display import display

# Open the image file
empty = Image.open('ARCANE_EMPTY.png')
full = Image.open('ARCANE_FULL.png')
full = full.convert('RGBA')

full_resize = full.resize((54, 54), Image.ANTIALIAS)

empty_arr = np.array(empty)
full_arr = np.array(full_resize)

empty_center = empty_arr[5:59, 5:59]

for i in range(54):
    mid = 27
    left = mid - i
    right = mid + i
    if left < 0:
        left = -left
    if right > 53:
        right = 53 - (right - 53)
    for j in range(54):
        if j < left or j > right:
            full_arr[i][j] = empty_center[i][j]
res_center = Image.fromarray(full_arr)
display(res_center)
res = empty_arr.copy()
res[5:59, 5:59] = full_arr
res = Image.fromarray(res)
display(res)
res.save('COST_ARCANE_FULL.png')
```
